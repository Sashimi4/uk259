# Task 2.1: Plotting

The variable *ts* below contains a randomly generated "time series" a dataset in which time is one of the axies.

* Calculate the **cumulative sum** of this series in a new series. The cumulative sum at index *i* is the sum of all entries in *ts* up until index *i*.
* Plot the cumulative sum using *matplotlib*

```python
import matplotlib.pyplot as plt
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts1 = ts.cumsum()
plt.plot(ts1)
```

# Which Seller sold the most houses ever

```python
import pandas as pd
import numpy as np
import scipy as scipy
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme(color_codes=True)

df = pd.read_csv("../datasets/melb_data.csv")



#df["SellerG"].value_counts()
#df["SellerG"].unique()

sellers = pd.DataFrame()
sellers["Amount Sold"] = df["SellerG"].value_counts()
sellers.index.name = "Seller names"
sellers.reset_index(inplace=True)
sellers.dropna(axis=1, how='any')
sellers.iloc[2:6]

#df[["Amount Sold", "SellerG"]]

# Top 17 Sellers
sns.barplot(x="Amount Sold", y="Seller names", data=sellers.iloc[2:17], color="salmon", estimator=np.median)
```

# Most wanted Suburb area

```python
import pandas as pd
import numpy as np
import scipy as scipy
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme(color_codes=True)

df = pd.read_csv("../datasets/melb_data.csv")

suburbs = pd.DataFrame()
suburbs["Amount sold in that suburb"] = df["Suburb"].value_counts()
suburbs.index.name = "Suburb name"
suburbs.reset_index(inplace=True)
suburbs.dropna(axis=1, how='any')
suburbs.iloc[2:6]

# Top 17 Sellers
sns.barplot(x="Amount sold in that suburb", y="Suburb name", data=suburbs.iloc[2:17], color="green", estimator=np.median)
```
